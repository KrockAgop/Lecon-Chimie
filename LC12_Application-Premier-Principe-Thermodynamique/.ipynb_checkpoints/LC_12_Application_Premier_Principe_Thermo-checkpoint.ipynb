```python
import numpy as np
import os
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
plt.rcParams["figure.figsize"]=10,5
```

```python
def my_round(X , DX) :
    Ndecimal = np.int(np.ceil(np.abs(np.log10(DX))))
    DX = np.around( DX , Ndecimal )
    X = np.around( X , Ndecimal )
    return X , DX

def affine(x,a,b) :
    return a * x + b

def lineaire(x,a) :
    return a * x
```
